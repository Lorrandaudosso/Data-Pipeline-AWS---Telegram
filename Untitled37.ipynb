## **O pipeline do projeto**

 Começa com a recepção de dados, que são armazenados no Amazon S3. Quando novos arquivos são adicionados, o AWS Lambda é acionado para processar esses dados. O Amazon EventBridge gerencia os eventos, garantindo que as funções Lambda sejam acionadas adequadamente para diferentes tipos de processamento.

Após o processamento, os dados são analisados pelo Amazon Athena, que permite executar consultas SQL diretamente nos dados armazenados no S3, gerando insights valiosos.

Finalmente, o Telegram Bot API é utilizado para interagir com os usuários, enviando notificações sobre atualizações ou resultados e recebendo comandos que podem iniciar novas ações. Esse pipeline integra armazenamento, processamento, análise e interação de forma contínua e automatizada, tornando o sistema eficiente e escalável.

**O API Gateway** é um serviço que atua como um ponto de entrada único para chamadas a uma API. Ele serve como um intermediário entre os clientes (como aplicativos web, dispositivos móveis ou sistemas internos) e um conjunto de microserviços ou outras APIs que compõem uma aplicação mais complexa.

**Um webhook** é uma maneira de permitir que um sistema envie informações automaticamente para outro sistema em tempo real, sem a necessidade de que o segundo sistema faça requisições constantes para verificar mudanças ou eventos (um processo conhecido como polling). O webhook é basicamente uma notificação automática que é enviada quando um determinado evento ocorre em um sistema, com os dados relevantes para o evento.

Para acionar o método setWebhook da API do Telegram Bot, você pode usar uma requisição GET. O método setWebhook é usado para configurar um webhook para o seu bot, permitindo que o Telegram envie atualizações (como mensagens) para um URL específico quando eventos ocorrerem.

### **AWS Athena**
Amazon Athena é um serviço de consulta interativa da Amazon Web Services (AWS) que permite que você analise dados diretamente no Amazon S3 utilizando SQL padrão. Ele é totalmente gerenciado e serverless (sem servidor), o que significa que você não precisa se preocupar com infraestrutura ou provisionamento de recursos. Athena é ideal para executar consultas em grandes volumes de dados armazenados em S3 sem precisar movê-los para um banco de dados ou data warehouse.

**Criando a tabeba no athena**
```sql
CREATE EXTERNAL TABLE `telegram`(
  `message_id` bigint,
  `user_id` bigint,
  `user_is_bot` boolean,
  `user_first_name` string,
  `chat_id` bigint,
  `chat_type` string,
  `text` string,
  `date` bigint)
PARTITIONED BY (
  `context_date` date)
ROW FORMAT SERDE
  'org.apache.hadoop.hive.ql.io.parquet.serde.ParquetHiveSerDe'
STORED AS INPUTFORMAT
  'org.apache.hadoop.hive.ql.io.parquet.MapredParquetInputFormat'
OUTPUTFORMAT
  'org.apache.hadoop.hive.ql.io.parquet.MapredParquetOutputFormat'
LOCATION
  's3://<bucket>/telegram/'
```

* Quantidade de mensagens por hora por dia da semana por número da semana.
```sql
WITH
parsed_date_cte AS (
    SELECT
        *,
        CAST(date_format(from_unixtime("date"),'%Y-%m-%d %H:%i:%s') AS timestamp) AS parsed_date
    FROM "telegram"
),
hour_week_cte AS (
    SELECT
        *,
        EXTRACT(hour FROM parsed_date) AS parsed_date_hour,
        EXTRACT(dow FROM parsed_date) AS parsed_date_weekday,
        EXTRACT(week FROM parsed_date) AS parsed_date_weeknum
    FROM parsed_date_cte
)
SELECT
    parsed_date_hour,
    parsed_date_weekday,
    parsed_date_weeknum,
    count(1) AS "message_amount"
FROM hour_week_cte
GROUP BY
    parsed_date_hour,
    parsed_date_weekday,
    parsed_date_weeknum
ORDER BY
    parsed_date_weeknum,
    parsed_date_weekday
```

# **Comclusão**

Este projeto demonstra a criação de um pipeline de dados altamente eficiente, integrando serviços da AWS com a API do Telegram. Utilizando serviços como AWS Lambda, S3, EventBridge e Athena, o sistema oferece uma solução escalável para coleta, armazenamento, processamento e análise de dados em tempo real, com notificações enviadas diretamente via Telegram.

Principais Pontos:
Automação: A combinação de EventBridge e Lambda garante a execução de funções automaticamente em resposta a eventos, como a adição de novos dados no S3 ou gatilhos programados, sem necessidade de intervenção manual.
Análise de Dados: O Athena facilita a consulta de dados diretamente no S3 utilizando SQL, permitindo análises rápidas sem mover os dados para um banco de dados tradicional.
Integração com Telegram: O uso da Telegram Bot API para envio de notificações e recebimento de comandos cria um fluxo contínuo de comunicação entre o pipeline e os usuários, tornando o sistema interativo.
Armazenamento Seguro: O Amazon S3 serve como o repositório central para dados crus e processados, garantindo durabilidade e acessibilidade global.
Benefícios:
Este pipeline pode ser utilizado em diversas aplicações, como monitoramento de sistemas, coleta de dados de usuários e até mesmo notificações personalizadas. A arquitetura serverless reduz a necessidade de gerenciamento de servidores, enquanto a integração com o Telegram torna a solução mais acessível e prática para os usuários finais.

Em suma, este projeto ilustra como a integração entre AWS e Telegram pode gerar um sistema dinâmico, automatizado e eficiente para coleta, análise e interação com dados em tempo real.
